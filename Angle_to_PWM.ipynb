```python
def angle_to_pulse(angle):
    A = 7.4074  # Slope based on your calibration
    B = 500     # Intercept at 0 degrees
    pwm_value = int(A * (angle) + B)
    return pwm_value
def add_ofset_to_angle(desired_angle):
  A = 0.051
  ofset = A*(desired_angle)
  return ofset

# Example usage
desired_angle = 180  # For example, moving to 90 degrees
desired_angle = desired_angle + add_ofset_to_angle(desired_angle)
pulse = angle_to_pulse(desired_angle)
print(f"Pulse Width for {desired_angle} degrees: {pulse} μs")
```

```python
import numpy as np

def convert(coxa,femur,tibia):
    pulse = np.empty([3])
    #FR
    pulse[0] = int(7.4074 * (-coxa)) + 1890
    pulse[1] = int(7.4074 * femur) + 524
    pulse[2] = int(7.4074 * tibia) + 570
    return pulse
desire_angles = np.array([83.10789742065361, 45.4595489556731, 138.78331244583745])
desire_angles = convert(desire_angles[0],desire_angles[1],desire_angles[2])
print(f"coxa.writeMicroseconds({desire_angles[0]});\n  femur.writeMicroseconds({desire_angles[1]});\n  tibia.writeMicroseconds({desire_angles[2]}); ")
```
